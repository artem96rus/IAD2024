#### Задание 1. Соберите всю информацию о студентах в одну таблицу df. В получившейся таблице должна быть информация и оценки всех студентов из всех групп. Напечатайте несколько строк таблицы для демонстрации результата.¶

```python
import zipfile
import os
```

```python
with zipfile.ZipFile('Data.zip', 'r') as zip_ref:
    zip_ref.extractall('data_students')

# Папка с разархивированными файлами
folder_path = 'data_students/Data/'

# Список для хранения данных из всех CSV
data_students = []
data_marks = []
for i in range(10):
  x = folder_path + 'Students_info_'+str(i)+'.csv'
  df_students = pd.read_csv(x)
  data_students.append(df_students)

  x = folder_path + 'Students_marks_'+str(i)+'.csv'
  df_marks = pd.read_csv(x)
  data_marks.append(df_marks)

df_students = pd.concat(data_students, ignore_index=True)
df_marks = pd.concat(data_marks, ignore_index=True)
df = pd.concat([df_students, df_marks], axis = 1)
df.head()
```
